```python
#Importamos las librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
# Importamos el dataset
df = pd.read_csv('covid.csv')
```

```python
# Nos familiarizamos con el dataset
df.head()
```

```python
df.describe()
```

```python
# Eliminamos columnas innecesarias y las que sobrepasan de 10 valores unicos
df.drop(['id','country','iso_code','date'],axis=1,inplace=True)
```

```python
# Hacemos el analisis de correlacion
corr = round(df.corr(),2)
corr.style.background_gradient()
```

```python
# Inicializamos el modelo
linear = linear_model.Ridge()
```

```python
# Preparamos la data
data_x = df.drop(['people_fully_vaccinated','New_deaths','ratio'], axis=1)
data_y = df.people_fully_vaccinated
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x,data_y,test_size=0.1)
```

```python
# Entrenamos el modelo
linear.fit(train_x,train_y)
```

```python
# Vemos el score de nuestro modelo
print(f'El score del modelo es: {round(linear.score(test_x,test_y)*100,2)}%')
```

```python
# Predecir
total_vaccinations = 12300474
people_vaccinated = 7462808
population = 19275294
prediccion = linear.predict([[total_vaccinations,people_vaccinated,population]])

# La prediccion correcta es 4837666
```

```python
print(f'Existe un total de {total_vaccinations} vacunas, \nlas personas que han asistido a la vacunacion son un total de {people_vaccinated} \ny la poblacion total es de {population}. \nLas personas que recibieron la dosis completa son: {round(prediccion[0],2)}')
```
