```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
# Cargamos el Dataset que utilizaremos
df = pd.read_csv('test.csv')
```

```python
# Eliminamos las columnas innecesarias
df.drop(['id', 'Customer Type', 'Type of Travel', 'Arrival Delay in Minutes',
                'Departure Delay in Minutes', 'Flight Distance', 'Inflight wifi service',
                'Departure/Arrival time convenient', 'Ease of Online booking',
                'Leg room service', 'Checkin service', 'Baggage handling',
                'On-board service', 'Unnamed: 0', 'Gate location'], axis='columns', inplace=True)
```

```python
df.head()
```

```python
# Mapeamos para tener valores enteros
df.Gender = df.Gender.map({'Male': 1, 'Female': 2})
df.Class = df.Class.map({'Business': 1, 'Eco': 2, 'Eco Plus': 3})
df.satisfaction = df.satisfaction.map({'satisfied': 1, 'neutral or dissatisfied': 2})
```

```python
# Separamos nuestros datos en características (x) y etiquetas (y)
data_x = df.drop('satisfaction', axis='columns')
data_y = df.satisfaction
```

```python
# Convertimos los datos a arrays de numpy
x = np.array(data_x)
y = np.array(data_y)
```

```python
# Normalizamos los valores de entrada
x = x / np.amax(x, axis=0)
```

```python
# Definimos la arquitectura de la red neuronal
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(11, activation='sigmoid'),
    tf.keras.layers.Dense(3, activation='sigmoid'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
# Compilamos el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Entrenamos el modelo sin mostrar el progreso de cada época
history = model.fit(x, y, epochs=10, verbose=0)
```

```python
# Evaluamos el modelo
loss, accuracy = model.evaluate(x, y, verbose=0)
print(f"La Tasa de Perdida fue: {round(loss, 2)}")
```

```python
# Precision de la red neuronal
print(f"Score: {round(accuracy, 2)}")
```

```python
# Imprimimos el número de épocas
numero_epocas = len(history.epoch)
print(f"El Número de Épocas fue: {numero_epocas}")
```
