```python
# Importando lo necesario de dependencias
from sklearn.model_selection import train_test_split
from sklearn import linear_model

import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# Importando csv y leyendo como DataFrame
df = pd.read_csv('car data.csv')
```

```python
df.head(10)
```

```python
df.Transmission.unique()
df.Seller_Type.unique()
df.Fuel_Type.unique()
```

```python
df.Fuel_Type = df['Fuel_Type'].map({'Petrol': 0, 'Diesel': 1, 'CNG': 2})
df.Seller_Type = df['Seller_Type'].map({'Dealer': 0, 'Individual': 1})
df.Transmission = df['Transmission'].map({'Manual': 0, 'Automatic': 1})
```

```python
df.describe()
```

```python
df.drop('Car_Name', axis='columns', inplace=True)
```

```python
corr = df.corr()
corr.style.background_gradient()
```

```python
# Inicializo el modelo
linear = linear_model.LinearRegression()
```

```python
# Preparo la data
data_x = df.drop(['Selling_Price', 'Seller_Type', 'Owner'], axis=1)
data_y = df.Selling_Price
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.1)
```

```python
# Entrenando el modelo
linear.fit(train_x, train_y)
```

```python
# Calculando ecuación de la recta
print(f'y = {round( linear.coef_[0], 2)}x + {round(linear.intercept_,2)}')
```

```python
# Viendo la calificación
print(f'Score: {round(linear.score(test_x, test_y), 4) * 100}%')
```

```python
# Vamos a predecir
year = 2008
present_price = 9.80
km_driven = 5100
fuel_type = 1
transmission_type = 0
predicted = linear.predict([[year, present_price, km_driven, fuel_type, transmission_type]])
labels_t = {0: 'Manual', 1: 'Automatico'}
labels_f = {0: 'Petrol', 1: 'Diesel', 2: 'CNG'}
```

```python
print(f'El auto del año {year}, con {km_driven} km de recorrido, de caja {labels_t[transmission_type]} y que usa {labels_f[fuel_type]} costaría {round(predicted[0],2)} Dolares aproximadamente.')
```
