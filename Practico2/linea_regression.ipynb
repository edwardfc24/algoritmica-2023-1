```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#aqui importo el dataset
df = pd.read_csv('desastresNaturales.csv')
```

```python
#aqui miro los campos y cantidad de datos
df.describe()
```

```python
#muestra los tipos de datos de los campos
df.dtypes
```

```python
# elimino estos campos innecesario
df = df.drop(['No Homeless','Associated Dis','Associated Dis2','Appeal','Declaration','Aid Contribution','River Basin','Local Time'], axis=1)
```

```python
# elimino estos campos innecesario
df = df.drop(['Dis No','Glide','ISO','Event Name'], axis=1)
```

```python
# elimino estos campos innecesario
df = df.drop(['CPI', 'Adm Level','Admin1 Code','Admin2 Code'], axis=1)
```

```python
#vemos los valores del campo 'Disaster Group'
gruop = df['Disaster Group'].unique()
gruop
```

```python
# llenar los valores vacios con 0
df['Disaster Group'].fillna(0, inplace=True)
```

```python
#mapear natural
df['Disaster Group'] = df['Disaster Group'].map({'Natural': 1})
```

```python
# vemos los valores del campo 'Disaster Subgroup'
Subgroup = df['Disaster Subgroup'].unique()
Subgroup
```

```python
df['Disaster Subgroup'] = df['Disaster Subgroup'].map({'Hydrological': 1,'Meteological':2,'Climatological':3,'Geophysical':4,
       'Biological':5, 'Extra-terrestrial': 6})
```

```python
type = df['Country'].unique()
type
```

```python
type = df['Origin'].unique()
type
```

```python
df['Disaster Type'] = df['Disaster Type'].map({'Flood':1, 'Storm':2, 'Drought':3, 'Landslide':4, 'Earthquake':5, 'Epidemic':6,
       'Wildfire':7, 'Mass movement (dry)':8, 'Extreme temperature':9,
       'Volcanic activity':10, 'Insect infestation':11, 'Impact':12,
       'Animal accident':13, 'Glacial lake outburst':14})
```

```python
 # eliminar 
df = df.drop(['Geo Locations'], axis=1)
```

```python
type = df['Region'].unique()
type
```

```python
type = df['Continent'].unique()
type
```

```python
df['Continent'] = df['Continent'].map({'Americas':1, 'Oceania':2, 'Africa':3, 'Asia':4, 'Europe':5})
```

```python
type = df['OFDA Response'].unique()
type
```

```python
df.dtypes
```

```python
df['OFDA Response'] = df['OFDA Response'].map({ 'Yes':1})
df['OFDA Response'].fillna(0, inplace=True)
```

```python
type = df['Dis Mag Scale'].unique()
type
```

```python
df['Dis Mag Scale'] = df['Dis Mag Scale'].map({'Km2':1, 'Kph':2, 'Richter':3, 'Vaccinated':4, '°C':5})
df['Dis Mag Scale'].fillna(0, inplace=True)
```

```python
df = df.drop(['Longitude','Latitude'], axis=1)
```

```python
df = df.drop(['Disaster Group'], axis=1)
```

```python
df.dtypes
```

regresion lineal

```python
df.describe()
```

```python
df = df.drop(['Disaster Subtype','Disaster Subsubtype','Country','Region','Location','Origin'], axis=1)
```

```python
#eliminar todos los nulos
df = df.dropna()
```

```python
df.dtypes
```

```python
# Análisis de correlación
corre = df.corr()
corre.style.background_gradient()
```

```python
linear = linear_model.LinearRegression()
```

```python
# nos quedamos con Insured Damages ('000 US$) y Reconstruction Costs ('000 US$)
data_x = df.drop(["Total Damages ('000 US$)",'Seq','Disaster Subgroup','Disaster Type','Continent','OFDA Response'
                  ,'Dis Mag Value','Dis Mag Scale','Start Year','Start Month','Start Day','End Year','End Month','End Day',
                  'No Affected','No Injured','Total Deaths','Total Affected', 'Year','Continent'], axis=1)
data_y = df["Total Damages ('000 US$)"]
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.5)
```

```python
# Entrenamos el modelo
linear.fit(train_x, train_y)
```

```python
print(f'y = {round( linear.coef_[0], 2)}x + {round(linear.intercept_, 2)}')
```

```python
print(f'Score: {round(linear.score(test_x, test_y), 4) * 100}%')
```

```python
reconstruction = 600000
insured = 2000
predicted = linear.predict([[reconstruction, insured]])
```

```python
predicted[0]
```
