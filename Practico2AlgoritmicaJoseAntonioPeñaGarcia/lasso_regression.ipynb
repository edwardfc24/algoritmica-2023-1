```python
# Importamos las librerias necesarias
from sklearn.model_selection import train_test_split
from sklearn import linear_model

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importamos el dataset para manejarlo como un DataFrame
df = pd.read_csv('emissions_cars.csv')
```

```python
#Aplicamos el describe para validar los datos
df.describe()
```

```python
df.drop(['Make', 'Model', 'Vehicle Class', 'Transmission', 'Fuel Type'], axis=1, inplace=True)
```

```python
#Analisis de correlacion 
corr = df.corr()
corr.style.background_gradient()
```

```python
# Inicializamos el modelo
# Otro ejemplo linear_model.ElasticNet()
Lasso = linear_model.Lasso()
```

```python
#Preparamos la data 
data_x = df.drop(['CO2 Emissions(g/km)','Fuel Consumption Comb (mpg)'], axis=1)
data_y = df['CO2 Emissions(g/km)']
```

```python
data_x
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.1)
```

```python
# Entrenamos el modelo
Lasso.fit(train_x, train_y)
# Lasso (operador de selección y contracción mínima absoluta) es un método de análisis de regresión que realiza selección de variables y regularización para mejorar la exactitud e interpretabilidad del modelo estadístico producido por este.
```

```python
# Veamos la calificacion
print(f'Score: {round(Lasso.score(test_x, test_y), 4) * 100}%')
```

```python
# Vamos a predecir
engine_size=2.4
cylinders=4
fuel_consuption_city=11.2
fuel_consuption_hwy=7.7
fuel_consuption_comb=9.6

predicted = Lasso.predict(X=[[engine_size, cylinders, fuel_consuption_city, fuel_consuption_hwy,fuel_consuption_comb]])
```
