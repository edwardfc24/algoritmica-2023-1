```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CO2 Emissions_Canada.csv')
```

```python
df = df.drop(['Make', 'Model', 'Vehicle Class', 'Transmission'], axis=1)
df['Fuel Type'] = df['Fuel Type'].map({'Z':0, 'X':1, 'E':2, 'D':3, 'N':4})
```

```python
df.describe()
```

```python
corr = df.corr()
corr.style.background_gradient()
```

```python
linear = linear_model.LinearRegression()
```

```python
data_x = df.drop(['Fuel Type', 'Fuel Consumption Comb (mpg)', 'Fuel Consumption Hwy (L/100 km)', 'Fuel Consumption Comb (L/100 km)', 'CO2 Emissions(g/km)'], axis=1)
data_y = df['CO2 Emissions(g/km)']
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.1)
```

```python
linear.fit(train_x, train_y)
```

```python
print(f'y = {round(linear.coef_[0], 2)}x + {round(linear.intercept_, 2)}')
```

```python
print(f'Score: {round(linear.score(test_x, test_y), 2) * 100}%')
```

```python
engine_size = 2.4
cylinders = 4
fuel_consumption_city = 9.9
predicted = linear.predict([[engine_size, cylinders, fuel_consumption_city]])
print(f'Un auto con tamaño de motor {engine_size}, {cylinders} cilindros y que consume {fuel_consumption_city} L/100 km, tendra una emision de CO2 de {round(predicted[0], 2)}g/km')
```
