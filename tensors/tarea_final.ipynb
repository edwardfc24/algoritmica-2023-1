```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import random_split, TensorDataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('fetal_health.csv')
```

```python
df.head()
```

```python
datos_y = df[df.columns[-1]]
datos_y.head()
```

```python
datos_x = df.drop(columns=["fetal_health"])
datos_x.head()
```

```python
datos_x = pd.get_dummies(datos_x)
datos_x.head()
```

```python
escalador = StandardScaler()
datos_x = escalador.fit_transform(datos_x)
print(datos_x)
```

```python
datos_x.shape[0]
```

```python
train_x, test_x, train_y, test_y = train_test_split(datos_x, datos_y, test_size=0.2, random_state=2)
```

```python
n_entradas = train_x.shape[1]
```

```python
t_train_x = torch.from_numpy(train_x).float().to("cpu")
t_test_x = torch.from_numpy(test_x).float().to("cpu")
t_train_y = torch.from_numpy(train_y.values).float().to("cpu")
t_test_y = torch.from_numpy(test_y.values).float().to("cpu")
t_train_y = t_train_y[:, None]
t_test_y = t_test_y[:, None]
```

```python
test = TensorDataset(t_test_x, t_test_y)
print(test[0])
```

```python
t_train_y
```

```python
class Red(nn.Module):
    def __init__(self, n_entradas):
        super(Red, self).__init__()
        self.linear1 = nn.Linear(n_entradas, 15)
        self.linear2 = nn.Linear(15, 8)
        self.linear3 = nn.Linear(8, 1)

    def forward(self, inputs):
        pred_1 = torch.sigmoid(input=self.linear1(inputs))
        pred_2 = torch.sigmoid(input=self.linear2(pred_1))
        pred_f = torch.sigmoid(input=self.linear3(pred_2))
        return pred_f
```

```python
%%time

lr = 0.001
epochs = 2000
status_print = 100

model = Red(n_entradas=n_entradas)
loss_fn = nn.BCELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
print("Arquitectura del modelo: {}".format(model))
historic = pd.DataFrame()

print("Entrenamiento del modelo")
for epoch in range(1, epochs+1):
    y_pred = model(t_train_x)
    loss = loss_fn(input=y_pred, target=t_train_y)
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()

    if epoch % status_print == 0:
        print(f"\nEpoch {epoch} \t Loss: {round(loss.item(), 4)}")

    with torch.no_grad():
        y_pred = model(t_test_x)
        y_pred_class = y_pred.round()
        correct = (y_pred_class == t_test_y).sum()
        accuracy = 100 * correct / float(len(t_test_y))
        if epoch % status_print == 0:
            print("Accuracy: {}".format(accuracy.item()))

    df_tmp = pd.DataFrame(data={
    "Epoch": [epoch],
    "Loss": [loss.item()],
    "Accuracy": [round(accuracy.item(), 4)]
    })
    historic = pd.concat(objs=[historic, df_tmp])
    
print("Accuracy final: {}".format(round(accuracy.item(), 4)))
```
