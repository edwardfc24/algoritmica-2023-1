```python
import pandas as pd;

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('dataset.csv')
```

```python
df.head()
```

```python
df.Target = df.Target.map({'Dropout' : 0, 'Graduate' : 1, 'Enrolled' : 2})
df.head()
```

```python
df.describe()
```

```python
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
# Curricular units 1st sem (enrolled)	Curricular units 1st sem (evaluations)	Curricular units 1st sem (approved)	Curricular units 1st sem (grade)	Curricular units 1st sem (without evaluations)	Curricular units 2nd sem (credited)	Curricular units 2nd sem (enrolled)	Curricular units 2nd sem (evaluations)	Curricular units 2nd sem (approved)	Curricular units 2nd sem (grade)	Curricular units 2nd sem (without evaluations)
data_x = df[[
    'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (evaluations)',
    'Curricular units 1st sem (grade)',
    'Curricular units 2nd sem (credited)',
    'Curricular units 2nd sem (enrolled)',
    'Curricular units 2nd sem (evaluations)',
    'Curricular units 2nd sem (approved)',
]]
data_y = df['Curricular units 2nd sem (grade)']
```

```python
train_x, test_x, train_y, test_y =  train_test_split(data_x, data_y, test_size=0.1)
```

```python
linear = DecisionTreeRegressor()
```

```python
linear.fit(train_x, train_y)
```

```python
linear.score(test_x, test_y)
```
