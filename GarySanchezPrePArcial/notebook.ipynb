```python
import pandas as pd
import matplotlib as plt
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
df.head()
```

```python
df['race/ethnicity'] = df['race/ethnicity'].map({'group A': 0, 'group B': 1, 'group C': 2, 'group D': 3, 'group E': 4})
df['race/ethnicity'].unique()

```

```python
df['parental level of education'] = df['parental level of education'].map({'some high school': 0, 'high school': 1, 'some college': 2, "associate's degree": 3, "bachelor's degree": 4, "master's degree": 5})
df['parental level of education'].unique()
```

```python
df['lunch'] = df['lunch'].map({'free/reduced': 0, 'standard': 1})
df['lunch'].unique()
```

```python
df['test preparation course'] = df['test preparation course'].map({'none': 0, 'completed': 1})
df['test preparation course'].unique()
```

```python
df['sex'] = df['sex'].map({'M': 0, 'F': 1})
```

```python
df.describe()
```

```python
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
k_neighbours = KNeighborsRegressor()
linear = LinearRegression()
```

```python
data_x = df[['reading score percentage', 'writing score percentage', 'lunch']]
data_y = df['math percentage']
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.1)
```

```python
linear.fit(train_x,train_y)
```

```python
print(f'Score Linear reg: {round(linear.score(test_x, test_y), 10) * 100}%')
```
