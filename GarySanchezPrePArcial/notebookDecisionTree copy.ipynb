```python
import pandas as pd
import matplotlib as plt
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv('data2.csv')
```

```python
df.head()
```

```python
df['parental level of education'] = df['parental level of education'].map({'some high school': 0, 'high school': 1, 'some college': 2, "associate's degree": 3, "bachelor's degree": 4, "master's degree": 5})
df.head()
```

```python
df.describe()
```

```python
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
decision_tree = DecisionTreeRegressor()
kneighbors = KNeighborsRegressor()
```

```python
data_x = df.drop(['years to graduate', 'parental level of education', 'parental income'], axis=1)
data_y = df['years to graduate']
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.1)
```

```python
decision_tree.fit(train_x, train_y)
```

```python
print(f'Score Linear reg: {round(decision_tree.score(test_x, test_y), 10) * 100}%')
```
