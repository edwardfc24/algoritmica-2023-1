```python
# Importar las librerías necesarias
from sklearn.model_selection import train_test_split
from sklearn import linear_model

import pandas as pd
import numpy as np
```

```python
# Importar el dataset y convertirlo en un DataFrame
df = pd.read_csv('Expanded_data_with_more_features.csv')
```

```python
df.head()
```

```python
df.drop(['Unnamed: 0', 'EthnicGroup', 'LunchType', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild', 'NrSiblings', 'TransportMeans'], axis=1, inplace=True)
```

```python
# Limpieza de los datos
test_preparations = {0: 'completó', 1: 'no completó'}
study_hours = {0: 'menos de 5', 1: 'entre 5 y 10', 2: 'más de 10'}

df.Gender = df.Gender.map({'female': 0, 'male': 1})
df.ParentEduc = df.ParentEduc.map({"bachelor's degree": 0, 'some college': 1, "master's degree": 2, "associate's degree": 3, 'high school': 4, 'some high school': 5})
df.TestPrep = df.TestPrep.map({'none': 0, 'completed': 1})
df.WklyStudyHours = df.WklyStudyHours.map({'< 5': 0, '5 - 10': 1, '> 10': 2})

df.ParentEduc = df.ParentEduc.fillna(df.ParentEduc.mean())
df.TestPrep = df.TestPrep.fillna(df.TestPrep.mean())
df.WklyStudyHours = df.WklyStudyHours.fillna(df.WklyStudyHours.mean())
```

```python
rows_without_nan = df.dropna(axis=0)
columns_without_nan = df.dropna(axis=1)
```

```python
# Analizar la correlación
corr = df.corr()
corr.style.background_gradient()
```

```python
# Inicializar el modelo
ridge = linear_model.Ridge()
```

```python
# Preparar la data
data_x = df.drop(['Gender', 'ParentEduc', 'WklyStudyHours', 'TestPrep', 'WritingScore'], axis=1)
data_y = df.WritingScore
```

```python
train_x, test_x, train_y, test_y = train_test_split(data_x, data_y, test_size=0.25)
```

```python
# Entrear el modelo
ridge.fit(train_x, train_y)
```

```python
# Calificación
print(f'score: {round(ridge.score(test_x, test_y), 4) * 100}%')
```

```python
# Predecir
math_score = 100
reading_score = 51
predicted = ridge.predict([[math_score, reading_score]])
```

```python
print(f'El estudiante que sacó {reading_score} en el examen de lectura y {math_score} en el examen de matemáticas, tendrá una nota aproximada de {round(predicted[0], 2)} en el examen de escritura')
```
